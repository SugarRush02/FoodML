```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\user\Desktop\ml\McDonalds.csv",index_col='Menu Items')
```

```python
df['Hunger Level']=0
df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X = df[['Energy (kCal)']] # Feature matrix
y = df['Hunger Level']    # Target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.92,random_state=40)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
y_train=[1,2,1,1,2,2,1,3,1,2,1]
print(y_train)
```

```python
lr = LinearRegression()

# Train the model on the training set
lr.fit(X_train, y_train)

# Predict the hunger level for the test set
y_pred = lr.predict(X_test)
```

```python
print('Coefficients:', lr.coef_)
print('Intercept:', lr.intercept_)
```

```python
print(X_test)
```

```python
y_pred=y_pred.round()
```

```python
y_pred=y_pred.astype('int')
```

```python
print(y_pred)
```

```python
df.loc[y_test.index,'Hunger Level']=y_pred
df.loc[X_train.index,'Hunger Level']=y_train
```
